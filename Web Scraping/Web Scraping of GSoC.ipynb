```python
import selenium 
from selenium import webdriver
import time
driver=webdriver.Chrome('./chromedriver')
driver.get('https://summerofcode.withgoogle.com/organizations/?sp-page=5#4590106242973696')
board=driver.find_elements_by_class_name('layout-wrap')
```

```python
cards=board[1].find_elements_by_class_name('organization-card__container')
```

```python
cards=board[1].find_elements_by_class_name('organization-card__container')
```

```python
organization_names=[]
for card in cards:
    name=card.find_element_by_class_name('organization-card__name')
    organization_names.append(name.text)
print(organization_names)    
```

```python

ul=driver.find_elements_by_tag_name('ul')
for l in ul:
    print(l.text)
    print()
```

```python
technologies=[]
topics_list=[]
for card in cards:
    card.click()
    tech=driver.find_elements_by_tag_name('ul')
    techs=tech[0].text.replace('\n',' , ')
    topics=tech[1].text.replace('\n',' ,  ')
    technologies.append(techs)
    topics_list.append(topics)
    time.sleep(1)
```

```python
print(technologies)
print(topics_list)
```

```python
import pandas as pd
```

```python
dataframe={'organization_name':organization_names, 'technologies':technologies, 'topics': topics_list}
```

```python
data=pd.DataFrame(dataframe)
```
